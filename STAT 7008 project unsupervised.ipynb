## 2. Models

### 2.1 Kmeans

###  &nbsp;&nbsp; 2.1.1 Fitting Models

```python
from sklearn.cluster import KMeans
import pandas as pd
kmeans = KMeans(n_clusters = 10).fit(x_trainf)
y_pred_kmean = kmeans.predict(x_testf)
```

###   &nbsp;&nbsp; 2.2.2 Clustering results evaluations

```python
test_score = kmeans.score(x_testf, y_test)
print('Test accuracy:', test_score)
```

#### &nbsp;&nbsp;&nbsp; NMI (Normaliszed Mutual Information)

```python
from sklearn.metrics.cluster import normalized_mutual_info_score
normalized_mutual_info_score(y_test, y_pred_kmean,average_method='arithmetic')
```

```python
from sklearn.metrics.cluster import normalized_mutual_info_score
normalized_mutual_info_score(y_test, y_pred_kmean,average_method='min')
```

```python
from sklearn.metrics.cluster import normalized_mutual_info_score
normalized_mutual_info_score(y_test, y_pred_kmean,average_method='geometric')
```

```python
from sklearn.metrics.cluster import normalized_mutual_info_score
normalized_mutual_info_score(y_test, y_pred_kmean,average_method='max')
```

####  &nbsp;&nbsp;&nbsp; ARI (Adjusted Rand Index)

```python
from sklearn.metrics.cluster import adjusted_rand_score
adjusted_rand_score(y_test, y_pred_kmean)
```

####   &nbsp;&nbsp;&nbsp; Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
kmeans_confusion =  confusion_matrix(y_test,y_pred_kmean)
print('Confusion matrix: \n{}'.format(kmeans_confusion))
```

```python
import seaborn as sns
sns.heatmap(kmeans_confusion, annot=True, cmap='Blues')
```

### 2.2 PCA (Principal component analysis)

###  &nbsp;&nbsp; 2.2.1 Scatter Plots in 2D and 3D

```python
from sklearn.decomposition import PCA

pca = PCA(n_components = 2, whiten = True)
pca.fit(x_trainf)
x_train_pca = pca.transform(x_trainf)
x_test_pca = pca.transform(x_testf)
```

```python
import matplotlib.pyplot as plt

%matplotlib inline
plt.scatter(x_train_pca[:,0], x_train_pca[:,1], c = y_train)
plt.show()
```

```python
pca3 = PCA(n_components = 3, whiten = True)
pca3.fit(x_trainf)
x_train_pca3 = pca3.transform(x_trainf)
x_test_pca3 = pca3.transform(x_testf)
```

```python
from mpl_toolkits.mplot3d import Axes3D
plt.figure(figsize=(8,8))
ax = plt.axes(projection="3d")
ax.scatter3D(x_train_pca3[:,0], x_train_pca3[:,1],x_train_pca3[:,2], c = y_train)
ax.view_init(10, 60)
```

```python
print(pca.explained_variance_ratio_)
print(pca3.explained_variance_ratio_)
```

###  &nbsp;&nbsp; 2.2.2 KMeans using PCA

```python
from sklearn.cluster import KMeans
import pandas as pd
kmeans = KMeans(n_clusters = 10).fit(x_train_pca)
y_pred_trans_kmean = kmeans.predict(x_test_pca)
test_score = adjusted_rand_score(y_test, y_pred_trans_kmean)
y_pred_kmean_pca = kmeans.predict(x_test_pca)
print('Test accuracy:', test_score)
```

####   &nbsp;&nbsp;&nbsp; Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
kmeans_pca_confusion =  confusion_matrix(y_test,y_pred_kmean_pca)
print('Confusion matrix: \n{}'.format(kmeans_pca_confusion))
```

```python
import seaborn as sns
sns.heatmap(kmeans_pca_confusion, annot=True, cmap='Blues')
```

### 2.3 MiniBatchKMeans

###  &nbsp;&nbsp; 2.3.1 Fitting Models

```python
from sklearn.cluster import MiniBatchKMeans
from sklearn.metrics.cluster import normalized_mutual_info_score
from sklearn.metrics.cluster import adjusted_rand_score
kmeans = MiniBatchKMeans(n_clusters=10,batch_size=5120).fit(x_train_pca)
y_pred_mbkm = kmeans.predict(x_test_pca)
```

####   &nbsp;&nbsp;&nbsp; Confusion Matrix

```python
mbkm_confusion = confusion_matrix(y_test,y_pred_mbkm)
print('Confusion matrix: \n{}'.format(mbkm_confusion))
```

```python
sns.heatmap(mbkm_confusion, annot=True, cmap='Blues')
```

### 3.2 Four Regression Models

```python
# Linear regression
from sklearn.linear_model import LinearRegression
lr = LinearRegression().fit(x_trainf,y_train)
print('Training data set score: {:.3f}'.format(lr.score(x_trainf, y_train)))
print('Test data set score: {:.3f}'.format(lr.score(x_testf, y_test))) # overfit
```

```python
# Ridge Regression
from sklearn.linear_model import Ridge
ridge = Ridge().fit(x_trainf, y_train)
print('Training data set score: {:.2f}'.format(ridge.score(x_trainf, y_train)))
print('Test data set score: {:.2f}'.format(ridge.score(x_testf, y_test))) # overfit
```

```python
# Lasso Regression
from sklearn.linear_model import Lasso
lasso = Lasso(alpha = 0.1, max_iter = 100000).fit(x_trainf, y_train)
print('Training data set score: {:.2f}'.format(lasso.score(x_trainf, y_train)))
print('Test data set score: {:.2f}'.format(lasso.score(x_testf, y_test))) # overfit
```

```python
# Logistic Regression
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression(max_iter=100).fit(x_trainf, y_train)
print('Training data set score: {:.3f}'.format(logreg.score(x_trainf, y_train)))
print('Test data set score: {:.3f}'.format(logreg.score(x_testf, y_test))) # overfit
```
