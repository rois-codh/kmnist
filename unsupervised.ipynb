## 1. Data and Clusters' Examples Visualization

```python
# download the dataset in NumPy format
import numpy as np
def load(f):
    return np.load(f)['arr_0']

# Load the data
x_train = load('kmnist-train-imgs.npz')
x_test = load('kmnist-test-imgs.npz')
y_train = load('kmnist-train-labels.npz')
y_test = load('kmnist-test-labels.npz')
```

```python
# Flatten images
# Each element in x_train and x_test is in the form of 28x28,
# so reshaping them in the form of 1x784
x_trainf = x_train.reshape(-1, 784)
x_testf = x_test.reshape(-1, 784)
```

## 2. Models

### 2.1 Kmeans

###  &nbsp;&nbsp; 2.1.1 Fitting Models

```python
# import KMeans from sklearn 
from sklearn.cluster import KMeans

# assigned 10 clusters 
# fit() to compute clustering for K-Means
# fit_predict() to make predictions
kmeans = KMeans(n_clusters = 10).fit(x_trainf)
y_pred_kmean = kmeans.predict(x_testf)
```

###   &nbsp;&nbsp; 2.2.2 Clustering results evaluations

#### &nbsp;&nbsp;&nbsp; NMI (Normaliszed Mutual Information)

```python
from sklearn.metrics.cluster import normalized_mutual_info_score
print("Test score:{:.4f}".format(normalized_mutual_info_score(y_test, y_pred_kmean,average_method='arithmetic')))
```

```python
from sklearn.metrics.cluster import normalized_mutual_info_score
print("Test score:{:.4f}".format(normalized_mutual_info_score(y_test, y_pred_kmean,average_method='min')))
```

```python
from sklearn.metrics.cluster import normalized_mutual_info_score
print("Test score:{:.4f}".format(normalized_mutual_info_score(y_test, y_pred_kmean,average_method='geometric')))
```

```python
from sklearn.metrics.cluster import normalized_mutual_info_score
print("Test score:{:.4f}".format(normalized_mutual_info_score(y_test, y_pred_kmean,average_method='max')))
```

####  &nbsp;&nbsp;&nbsp; ARI (Adjusted Rand Index)

```python
from sklearn.metrics.cluster import adjusted_rand_score
print("Test score:{:.4f}".format(adjusted_rand_score(y_test, y_pred_kmean)))
```

####   &nbsp;&nbsp;&nbsp; Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
kmeans_confusion =  confusion_matrix(y_test,y_pred_kmean)
print('Confusion matrix: \n{}'.format(kmeans_confusion))
```

```python
# visualize the confusion matrix, the deeper color represents the greater values
import seaborn as sns
sns.heatmap(kmeans_confusion, annot=False, cmap='Blues')
plt.show()
```

### 2.2 PCA (Principal component analysis)

###  &nbsp;&nbsp; 2.2.1 Scatter Plots in 2D and 3D

```python
# import PCA from sklearn
from sklearn.decomposition import PCA

# implement PCA and keep the first two principal components only
pca = PCA(n_components = 2, whiten = True)
pca.fit(x_trainf)

# transform data to reduce dimensions
x_train_pca = pca.transform(x_trainf)
x_test_pca = pca.transform(x_testf)
```

```python
import matplotlib.pyplot as plt

%matplotlib inline
plt.scatter(x_train_pca[:,0], x_train_pca[:,1], c = y_train)
plt.show()
```

```python
# implement PCA and keep the first three principal components only
pca3 = PCA(n_components = 3, whiten = True)
pca3.fit(x_trainf)
# transform data
x_train_pca3 = pca3.transform(x_trainf)
x_test_pca3 = pca3.transform(x_testf)
```

```python
from mpl_toolkits.mplot3d import Axes3D
plt.figure(figsize=(8,8))
ax = plt.axes(projection="3d")
ax.scatter3D(x_train_pca3[:,0], x_train_pca3[:,1],x_train_pca3[:,2], c = y_train)
ax.view_init(10, 60)
```

```python
print(pca.explained_variance_ratio_)
print(pca3.explained_variance_ratio_)
```

### 3.2 Four Regression Models

```python
# Linear regression
from sklearn.linear_model import LinearRegression
lr = LinearRegression().fit(x_trainf,y_train)
print('Training data set score: {:.3f}'.format(lr.score(x_trainf, y_train)))
print('Test data set score: {:.3f}'.format(lr.score(x_testf, y_test))) # overfit
```

```python
# Ridge Regression
from sklearn.linear_model import Ridge
ridge = Ridge().fit(x_trainf, y_train)
print('Training data set score: {:.2f}'.format(ridge.score(x_trainf, y_train)))
print('Test data set score: {:.2f}'.format(ridge.score(x_testf, y_test))) # overfit
```

```python
# Lasso Regression
from sklearn.linear_model import Lasso
lasso = Lasso(alpha = 0.1, max_iter = 100000).fit(x_trainf, y_train)
print('Training data set score: {:.2f}'.format(lasso.score(x_trainf, y_train)))
print('Test data set score: {:.2f}'.format(lasso.score(x_testf, y_test))) # overfit
```

```python
# Logistic Regression
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression(max_iter=100).fit(x_trainf, y_train)
print('Training data set score: {:.3f}'.format(logreg.score(x_trainf, y_train)))
print('Test data set score: {:.3f}'.format(logreg.score(x_testf, y_test))) # overfit
```
